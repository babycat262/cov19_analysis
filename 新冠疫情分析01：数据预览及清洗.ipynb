#### 城市 `cities` 的处理

```python
area_df[area_df['cities'].notnull()]
```

```python
def clean_cities(x):
    x = str(x).strip()
    if x in {'', ' ', np.nan}:
        return '暂无'
    elif 'None' in x or '[]' in x: 
        return '暂无'
    else:
        return x

area_df['cities'] = area_df['cities'].apply(clean_cities)
# check
area_df['cities'].value_counts()
```

```python
sum(area_df['cities'].isnull())
```

#### `comment` 列的整理

```python
area_df['comment'].value_counts()
```

```python
area_df[area_df['comment'].isnull()]
```

```python
# comment_map = {'': '暂无', ' ': '暂无', 'None': '暂无', None: '暂无' }
# area_df.comment.fillna('暂无', inplace=True)
area_df['comment'] = area_df['comment'].apply(lambda x: '暂无' if x in {'', ' ', 'None', None, np.nan} else x)
area_df['comment'].value_counts()
```

#### `countryEnglishName` 的处理

```python
area_df[area_df['countryEnglishName'].isnull()].index
```

```python
# 填充缺失值，以 '暂无' 代替
area_df['countryEnglishName'].fillna(value='暂无', inplace=True)
# check 
area_df[area_df['countryEnglishName']=='暂无'].index
```

#### `provinceEnglishName`的处理

```python
area_df[area_df['provinceEnglishName'].isnull()].index
```

```python
## 填充缺失值
area_df['provinceEnglishName'].fillna(value='暂无', inplace=True)
# check
area_df[area_df['provinceEnglishName']=='暂无'].index
```

```python
## 计算死亡率（计算结果保留4位小数）： 死亡率 = 死亡数 / 累计确诊数
area_df['deadRate'] = (area_df['deadCount'] / area_df['confirmedCount']).apply(lambda x: round(x, 4))
# check
area_df['deadRate'].head()
```

```python
# 计算治愈率（计算结果保留4位小数）： 治愈率 = 治愈人数 / 累计确诊人数
area_df['curedRate'] = (area_df['curedCount'] / area_df['confirmedCount']).apply(lambda x: round(x, 4))
# check
area_df['curedRate'].head()
```
